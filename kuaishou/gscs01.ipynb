```python
from PIL import Image

# 打开图片
image = Image.open('7201.png')

# 创建一个和原图大小相同的透明图像
overlay = Image.new('RGBA', image.size, (0, 0, 0, 0))

# 创建一个可绘制对象
draw = ImageDraw.Draw(overlay)

# 定义区域的位置和大小
x = 100
y = 100
width = 200
height = 200

# 定义区域的颜色和透明度
fill_color = (255, 255, 255, 128)  # 最后一个参数128表示50%的透明度

# 在透明图像上绘制半透明区域
draw.rectangle((x, y, x + width, y + height), fill=fill_color)

# 将透明图像与原图像混合
result = Image.alpha_composite(image.convert('RGBA'), overlay)

# 保存修改后的图片
result.save('D:/31/new.png')
```

```python

import requests

def get_poem():
    url = "https://v1.jinrishici.com/all.json"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        poem_data = response.json()
        return poem_data
    else:
        print("Error, status code:", response.status_code)
        return None

poem = get_poem()
print(poem)
if poem:
    print("标题：", poem["origin"])
    print("作者：", poem["author"])
    print("内容：", poem["content"])
```
