```python

import requests

def get_poem():
    url = "https://v1.jinrishici.com/all.json"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        poem_data = response.json()
        return poem_data
    else:
        print("Error, status code:", response.status_code)
        return None

poem = get_poem()
print(poem)
if poem:
    print("标题：", poem["origin"])
    print("作者：", poem["author"])
    print("内容：", poem["content"])
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import os

# 读取原始图片和水印图片
for i0 in range(500,600):
    img = Image.open('D:/1/bj2.jpg')
    wm1=f'D:\\onedriver1t\\OneDrive - 356h31\\100\\2-1\\a{str(i0)}.png'
    
    #判断文件是否存在,不存在跳过本循环
    if not os.path.exists(wm1):
        print('文件不存在')
        continue
    wm = Image.open(wm1)

   
   
    # 调整水印图片的大小和透明度
    wm_scale = 1 # 水印图片缩放比例
    wm_alpha = 1 # 水印图片透明度
    wm_width = int(wm.size[0] * wm_scale)
    wm_height = int(wm.size[1] * wm_scale)
    wm_dim = (wm_width, wm_height)
    wm = wm.resize(wm_dim, Image.ANTIALIAS)
    wm = wm.convert('RGBA') # 添加alpha通道
    r, g, b, a = wm.split() # 分离各个通道
    a = a.point(lambda i: i * wm_alpha) # 修改alpha通道的值
    wm.putalpha(a) # 重新合并各个通道

    # 将水印图片叠加到原始图片的右下角
    img = img.convert('RGBA') # 添加alpha通道
    w_img, h_img = img.size
    w_wm, h_wm = wm.size
    top_y = 50 # 水印图片的上边界坐标，10为边距
    left_x =  10 # 水印图片的左边界坐标，10为边距
    img.paste(wm, (left_x, top_y), wm) # 将水印图片粘贴到原始图片上，使用水印图片的alpha通道作为掩码

    # 保存并显示加水印后的图片
    img.save(f'D:\\3\\a{str(i0)}.png')
    # plt.imshow(img)
    # plt.show()
```
