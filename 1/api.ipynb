```python
import requests

url = "https://yf.sgai.cc/gptcms/api/rolechat/send"
headers = {
    "Wid": "1",
    "Token": "25d28cca-aea3-11ee-9f65-0242d1ee50c3"
}
params = {
    "model_id": 11,
    "message": "科幻故事"
}

response = requests.get(url, headers=headers, params=params)
data = response.text.split("\n")
data = [line for line in data if line.startswith("data:")]
data = [line.replace("data:", "").strip() for line in data]

print(data)
```
