```python
import openpyxl
from openpyxl.styles import Font, PatternFill, Border, Side, Alignment

# 创建一个新的工作簿和工作表
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "漂亮的表格"

# 添加表头
headers = ["姓名", "年龄", "性别", "分数"]
ws.append(headers)

# 添加数据
data = [
    ["张三", 23, "男", 88],
    ["李四", 22, "女", 95],
    ["王五", 24, "男", 78],
]

for row in data:
    ws.append(row)

# 设置字体和对齐方式
header_font = Font(bold=True, color="FFFFFF")
header_fill = PatternFill(start_color="4F81BD", end_color="4F81BD", fill_type="solid")
alignment = Alignment(horizontal="center", vertical="center")

# 设置边框
thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)

# 应用样式到表头
for col in ws.iter_cols(min_row=1, max_row=1, min_col=1, max_col=len(headers)):
    for cell in col:
        cell.font = header_font
        cell.fill = header_fill
        cell.alignment = alignment
        cell.border = thin_border

# 应用样式到数据单元格
for row in ws.iter_rows(min_row=2, max_row=2+len(data), min_col=1, max_col=len(headers)):
    for cell in row:
        cell.alignment = alignment
        cell.border = thin_border

# 调整列宽
for col in ws.columns:
    max_length = 0
    column = col[0].column_letter  # 获取列字母
    for cell in col:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2)
    ws.column_dimensions[column].width = adjusted_width

# 保存文件
wb.save("formatted_table.xlsx")
```

```python
import matplotlib.pyplot as plt  
  
# 假设数据  
cities = ['济南', '青岛', '烟台', '潍坊', '临沂', '济宁', '淄博', '东营', '泰安', '威海', '日照', '滨州', '德州', '聊城', '菏泽', '枣庄']  
gdp = [12757.4, 15760.34, 10162.46, 7606.01, 6105.17, 5516.47, 4561.79, 3899.06, 3323.86, 3513.54, 2390.86, 3118.85, 3805.27, 2926.36, 4464.49, 2156.74]  
  
# 绘制柱状图  
plt.figure(figsize=(10, 8))  # 设置图表大小  
plt.bar(cities, gdp, color='skyblue')  # 绘制柱状图  
plt.xlabel('城市')  # 设置x轴标签  
plt.ylabel('GDP (亿元)')  # 设置y轴标签  
plt.title('2023年山东各市GDP情况')  # 设置图表标题  
plt.xticks(rotation=45, ha='right')  # 旋转x轴标签，避免重叠  
plt.tight_layout()  # 自动调整子图参数, 使之填充整个图像区域  
plt.show()  # 显示图表
```
