```python
import requests
from bs4 import BeautifulSoup

def get_weibo_hot_search():
    url = "https://s.weibo.com/top/summary"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        hot_search_items = soup.find_all('tr', attrs={'action-type': 'hover'})

        hot_search_list = []
        for item in hot_search_items:
            rank = item.find('td', class_='td-01').text.strip()
            title = item.find('td', class_='td-02').a.text.strip()
            hot_degree = item.find('td', class_='td-03').text.strip()
            hot_search_list.append((rank, title, hot_degree))

        return hot_search_list
    else:
        print("Error: Failed to fetch the weibo hot search.")
        return None

if __name__ == "__main__":
    hot_search_list = get_weibo_hot_search()
    if hot_search_list:
        for rank, title, hot_degree in hot_search_list:
            print(f"{rank}. {title} - {hot_degree}")
```

```python
from PIL import Image, ImageDraw, ImageFont

# 打开图片
image = Image.open('1.png')

# 创建 Draw 对象
draw = ImageDraw.Draw(image)

# 定义字体和文字
font = ImageFont.truetype('msyh.ttc', 24)
text = '这是要添加的文字\n这是第二行\n这是第三行'

# 计算文字位置
text_width, text_height = draw.textsize(text, font=font)
x = (image.width - text_width) / 2
y = (image.height - text_height) / 2

# 添加文字
for line in text.split('\n'):
    line_width, line_height = draw.textsize(line, font=font)
    draw.text(((image.width - line_width) / 2, y), line, fill=(255, 255, 255), font=font)
    y += line_height

# 保存图片
image.save('2.png')
```
